# 爬取网站验证码

```python
from curl_cffi import requests
import os
import time
from datetime import datetime
from requests.packages.urllib3.exceptions import InsecureRequestWarning


def download_image(session, save_path):
    """
    下载图片并保存到指定路径
    """
    url = "https://corpebankq.psbc.com/eweb/GenTokenImg.do"
    try:
        response = session.get(url, verify=False)  # verify=False 跳过SSL验证
        if response.status_code == 200:
            with open(save_path, 'wb') as f:
                f.write(response.content)
            return True
        return False
    except Exception as e:
        print(f"下载失败: {str(e)}")
        return False

def main(interval=10, max_requests=5):
    """
    主控制函数
    :param interval: 请求间隔时间（秒）
    :param max_requests: 最大请求次数
    """
    # 创建存储目录
    save_dir = '../data'
    os.makedirs(save_dir, exist_ok=True)
    
    # 配置请求头
    headers = {
    "Accept": "*/*",
    "Accept-Encoding": "gzip, deflate, br",
    "User-Agent": "PostmanRuntime-ApipostRuntime/1.1.0",
    "Connection": "keep-alive"
}
    
    # 初始化会话
    session = requests.Session()
    session.headers.update(headers)
    
    for count in range(1, max_requests+1):
        # 生成唯一文件名
        timestamp = datetime.now().strftime("%Y%m%d%H%M%S%f")[:-3]
        filename = f"image_{timestamp}.png"
        save_path = os.path.join(save_dir, filename)
        
        
        # 执行下载
        print(f"正在执行第 {count}/{max_requests} 次请求...")
        download_image(session, save_path)
        
        # 最后一次不等待
        if count < max_requests:
            time.sleep(interval)
    
    print("所有任务执行完成")

    # 参数配置（按需修改）
main(interval=2, max_requests=1800)
```
